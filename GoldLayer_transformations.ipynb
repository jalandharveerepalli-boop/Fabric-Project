### GOLD LAYER TRANSFORMATIONS & AGGREGATIONS: SHOPPING MART DATA

```python
from pyspark.sql.functions import *

df_orders = spark.read.parquet("Files/ShoppingMart_Silver_Orders/ShoppingMart_Orders.parquet")
```

```python
df_weblogs = spark.read.parquet("Files/ShoppingMart_Silver_Weblogs/ShoppingMart_Weblogs.parquet")
df_weblogs = df_weblogs.groupBy("user_id","page","action").count()

df_weblogs.write.mode("overwrite").parquet("Files/ShoppingMart_Gold_Weblogs/ShoppingMart_Weblogs.parquet")
```

```python
df_social = spark.read.parquet("Files/ShoppingMart_Silver_Social/ShoppingMart_Social.parquet")
display(df_social)
df_social = df_social.groupBy("platform","sentiment").count()

#df_social.write.mode("overwrite").parquet("Files/ShoppingMart_Gold_Social/ShoppingMart_Social.parquet")
```

```python
df_reviews = spark.read.parquet("Files/ShoppingMart_Silver_Reviews/ShoppingMart_Reviews.parquet")
df_reviews = df_reviews.groupBy("product_id").agg(avg("rating").alias("avg_rating"))

df_reviews.write.mode("overwrite").parquet("Files/ShoppingMart_Gold_Reviews/ShoppingMart_Reviews.parquet")
```
