### SILVER Layer: Data Cleaning & Ingestion

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

### Reading Structured Data

```python
df_customers = spark.read.format("csv").option("header","true").load("Files/ShoppingMart_Bronze_customers/ShoppingMart_customers.csv")
df_orders = spark.read.format("csv").option("header","true").load("Files/ShoppingMart_Bronze_Orders/ShoppingMart_orders.csv")
df_products = spark.read.format("csv").option("header","true").load("Files/ShoppingMart_Bronze_products/ShoppingMart_products.csv")
```

```python
df_orders = df_orders.dropna(subset= ["OrderID","CustomerID","ProductID","TotalAmount"])
df_orders = df_orders.withColumn("orderDate", to_date(col("OrderDate")))
```

```python
df_orders = df_orders.join(df_customers, on= "CustomerID", how= "inner")\
                     .join(df_products, on = "ProductID", how= "inner")
```

### Reading Unstructured Data

```python
df_reviews = spark.read.format("json").load("Files/ShoppingMart_Bronze_reviews/ShoppingMart_reviews.json")
df_social = spark.read.format("json").load("Files/ShoppingMart_Bronze_social_media/ShoppingMart_social_media.json")
df_weblogs = spark.read.format("json").load("Files/ShoppingMart_Bronze_web_logs/ShoppingMart_web_logs.json")
```

```python
# convert timestamp to datatime format
df_reviews = df_reviews.withColumn("timestamp", col("timestamp")/1000)\
                       .withColumn("timestamp", to_timestamp("timestamp"))

display(df_reviews)
```

```python
df_weblogs = df_weblogs.withColumn("timestamp", col("timestamp")/1000)\
                       .withColumn("timestamp", to_timestamp("timestamp"))
display(df_weblogs)
```

### Writing Transformed data to Silver Layer as a Parquet

```python
df_orders.write.mode("overwrite").parquet("Files/ShoppingMart_Silver_customers/ShoppingMart_Orders.parquet")
```

```python
df_reviews.write.mode("overwrite").parquet("Files/ShoppingMart_Silver_reviews/ShoppingMart_Reviews.parquet")
df_social.write.mode("overwrite").parquet("Files/ShoppingMart_Silver_social/ShoppingMart_Social.parquet")
df_weblogs.write.mode("overwrite").parquet("Files/ShoppingMart_Silver_weblogs/ShoppingMart_Weblogs.parquet")
```
